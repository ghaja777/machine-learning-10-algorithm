# PCA 1

```python
from sklearn.preprocessing import StandardScaler

```

```python
df.head(10)
```

```python
feature=['patient_number','cholesterol','glucose','hdl_chol','chol_hdl_ratio','age','gender','height','weight','bmi','systolic_bp','diastolic_bp','waist','hip','waist_hip_ratio','diabetes']
```

```python
X=df.loc[:,feature]  #X = df[:,0:4]
y=df.loc[:,'diabetes'] # y =[:,4]
```

```python
X=StandardScaler().fit_transform(X)
X
```

```python
#Converting PCA excecution and projection to 2D
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pct=pca.fit_transform(X)# pct - model

principal_df=pd.DataFrame(pct,columns=['PC1','PC2'])

final_df=pd.concat([principal_df,df[['diabetes']]],axis=1)
final_df
```

```python
print('before pca\n++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
print(df[:5])
print('after pca\n++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
print(final_df[:5])
```

```python
principal_df
```

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('Two Component PCA')

targets=['gender','glucose','bmi']
colors=['b','g','c']

for target,color in zip(targets,colors):
    index=final_df['diabetes']==target
    ax.scatter(final_df.loc[index,'PC1'],final_df.loc[index,'PC2'],c=color,s=5)
    ax.legend(targets)
    #ax.grid()
plt.plot(X,y)

pca.explained_variance_ratio_
```

```python
from xgboost import XGBRegressor
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()
x,y = diabetes.data,diabetes.target
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.15)

xgbr = XGBRegressor(verbosity=0)
print(xgbr)
```

```python
xgbr.fit(xtrain,ytrain)
score=xgbr.score(xtrain,ytrain)
print('training score is:',score)
```

```python
cv_score = cross_val_score(xgbr,xtrain,ytrain,cv=10)
print('cv mean score is:',cv_score.mean())
```

```python
ypred=xgbr.predict(xtest)
mse=mean_squared_error(ytest,ypred)
print('MSE is:',mse)
print('rmse is:',mse*(1/2.0))
```

```python
x_ax=range(len(ytest))
plt.plot(x_ax,ytest,label='original')
plt.plot(x_ax,ypred,label='predicted')
plt.title('diabetes data actual vs predicted')
plt.legend()
plt.show()
```
